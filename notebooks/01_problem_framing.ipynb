# Phase 1: Problem Framing

## 1.1 Selected Problem: Migration Intelligence Tool

**Problem Statement:**

Using Aadhaar enrolment, demographic updates, and biometric update trends to detect population movement signals indicating urbanisation and regional migration.


## 1.2 Why this problem?

- UIDAI data is naturally time-series and suitable for trend analysis.
- Analysis is possible from State → District level with policy relevance.
- The problem supports urban planning, governance, and service delivery.
- Entire analysis is aggregate-level, ensuring privacy.

## 1.3 What exactly we will measure?

We are not tracking individuals. We are tracking aggregate behaviour shifts.

We aim to detect:
1. Sudden enrolment spikes in districts.
2. Age-group skews (18+) indicating working population movement.
3. Biometric update surges indicating settling workforce.
4. Consistent demographic update patterns over time.

## 2.1 Granularity Decision

We will analyse data at the following levels:

- **Primary Unit:** District level  
  (Most stable unit for migration and urbanisation signals)

- **Secondary Unit:** State level  
  (Used for validation and baseline comparison)

- **Optional Unit:** Pincode level  
  (Only for deep-dive, not for core signals to avoid noise)


## 2.2 Time Aggregation Strategy

- All daily records will be aggregated to **monthly totals**
- Daily data is noisy (holidays, operational variation)
- Monthly aggregation gives policy-relevant trends

**Time format:** YYYY-MM


## 3. Signal Blueprint (What we will compute later)

We will construct **district-month level signals** from UIDAI datasets.

### Core Signals

1. **Enrolment Growth Signal**
   - Metric: Month-on-month percentage change in total enrolments
   - Interpretation: New population inflow

2. **Adult Share Shift Signal**
   - Metric: Ratio of 18+ enrolments to total enrolments
   - Interpretation: Working-age migration

3. **Biometric Update Pressure**
   - Metric: Monthly biometric updates per 1000 enrolments
   - Interpretation: Settling workforce requiring authentication

4. **Demographic Update Stability**
   - Metric: Consistency of demographic updates across months
   - Interpretation: Address and identity stabilisation

### Output Unit
Each signal will be computed at:
**District × Month**
